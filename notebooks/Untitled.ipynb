```python
coco ={
  "images": [
    {
      "id": 1,
      "file_name": "img_001.jpg",
      "width": 640,
      "height": 480
    }
  ],  "categories": [
    {
      "id": 1,
      "name": "bottle",
      "supercategory": "container"
    },
    {
      "id": 2,
      "name": "cup",
      "supercategory": "container"
    },
    {
     "id": 3,
     "name": "bag",
     "supercategory": "container"
    },
    {
     "id": 4,
     "name": "shoe",
     "supercategory": "container"
    },
    {
     "id": 5,
     "name": "shirt",
     "supercategory": "container"
    }
      
  ],
  "annotations": [
    {
      "id": 10,
      "image_id": 1,
      "category_id": 1,
      "bbox": [120, 80, 200, 300],
      "area": 60000,
      "iscrowd": 0
    }
  ]
}
```

```python
categories = {}
for cat in coco["categories"]:
    categories[cat["id"]] = cat["name"]
    
```

```python
cate = {cat["id"]: cat["name"] for cat in coco["categories"]}
cate
```

```python
category_ids = sorted(cate.keys())
category_ids
```

```python
category_names = [cate[cid] for cid in category_ids]
category_names
```

```python
for cid in category_ids:
    print(cate[cid])
```

```python
image_annotation = {}
for ann in coco["annotations"]:
    image_id = ann["image_id"]
    if image_id not in image_annotation:
        image_annotation[image_id] = []
    image_annotation[image_id].append(ann)
```
